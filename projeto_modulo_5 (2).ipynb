## Análise das Razões de Saída dos Funcionários: Comparação entre Percepção Interna e Realidade - Um Estudo de Caso no Departamento de Marketing


- # Evidência: Discrepâncias nas Percepções de Saída dos Funcionários

Neste projeto, realizei uma análise minuciosa das razões pelas quais os funcionários estão deixando a empresa, com foco especial no Departamento de Marketing. O cerne da investigação foi a identificação e compreensão das discrepâncias entre as percepções internacionais da organização, representadas pelo Parceiro de Negócios de Recursos Humanos (HRBP), e a realidade dos motivos subjacentes à rotatividade.

Para atingir esse objetivo, eu utilizei técnicas de visualização de dados avançadas, incluindo a criação de um gráfico de bolhas e   gráficos de barras  que permitiu a representação clara das razões de saída em relação à percepção da empresa e aos dados da pesquisa de saída. Ao comparar os dois conjuntos de dados, eu identifiquei áreas de desconexão e discrepâncias notáveis.

As escolhas feitas durante o projeto, desde a seleção das informações relevantes até a criação de gráficos informativos, foram orientadas pelo objetivo de fornecer uma análise precisa e acionável. Optei por usar um gráfico de bolhas e gráficos de barras para visualizar as discrepâncias, pois essas abordagens gráficas permite uma representação clara e comparativa das percepções internas e dos dados reais de saída.

Acredito que essa abordagem ofereceu uma visão significativa das questões subjacentes à rotatividade no Departamento de Marketing e na organização como um todo. Além disso, essa escolha de destacar as discrepâncias ressalta a importância de uma comunicação transparente e do desenvolvimento de estratégias de retenção de talentos mais alinhados com as preocupações reais dos funcionários.

No contexto deste projeto, a evidência apresentada não apenas fornece uma compreensão mais profunda das discrepâncias nas percepções de saída dos funcionários, mas também defende uma abordagem exigida para atingir esse objetivo, garantindo que as conclusões sejam baseadas em dados sólidos e análises rigorosas.

- Explicação do Gráfico de Bolhas do HRBP (Parceiro de Negócios de Recursos Humanos)

Este é o gráfico de bolhas que representa a percepção interna da empresa, especificamente a visão do HRBP (Parceiro de Negócios de Recursos Humanos), sobre as razões pelas quais os funcionários estão deixando a organização. Aqui está como ele funciona:

Bolhas Coloridas :

Cada bolha representa colorida uma razão específica de saída.
O tamanho de cada bolha reflete o quão importante o HRBP percebe essa razão como um fator que leva os funcionários a deixar a empresa.
Bolhas maiores indicam que o HRBP acredita que essa razão tem um impacto mais significativo na saída dos funcionários.


- Explicação do Gráfico de Bolhas do Exit Survey

Este é o gráfico de bolhas que representa os dados reais encontrados por meio da pesquisa de saída, fornecendo uma visão objetiva das razões pelas quais os funcionários estão deixando a empresa. Aqui está como ele funciona:

Bolhas em Toneladas Mais Claros :

Cada bolha em tom mais claro representa uma razão específica de saída, correspondente à mesma razão nos gráficos do HRBP.
O tamanho de cada bolha reflete a frequência com que essa razão foi mencionada pelos funcionários que participaram da pesquisa de resultado.
Bolhas maiores indicam que essa razão foi citada com mais frequência pelos funcionários como um motivo para deixar a empresa.


Comparando os dois gráficos, identifiquei discrepâncias notáveis ​​entre a percepção interna da empresa (HRBP) e a realidade dos motivos da rotatividade conforme relatados pelos funcionários na pesquisa de resultado.
Se uma bolha estiver posicionada mais alta nos gráficos do HRBP, mas a mesma razão por ser representada por uma bolha menor nos dados da pesquisa de saída, isso indica uma discrepância.
Essas discrepâncias são áreas em que a empresa pode não estar totalmente ciente das verdadeiras preocupações dos funcionários que estão saindo.

- ***vantagens e insuficiências de cada uma.***

1. Gráfico de Barras com Média:

Vantagens:

Simplicidade : Os gráficos de barras são fáceis de entender e interpretar, tornando-os acessíveis para uma ampla audiência.


Médias Descritivas : Usar uma mídia pode fornecer uma visão geral das percepções médias da empresa (HRBP) e da pesquisa de saída em relação a cada motivo de saída.
Comparação Simplificada : Permite uma comparação direta entre as médias das duas fontes de dados.

Limitações:

Perda de Detalhes : Uma mídia pode obscurecer variações e nuances nos dados. Pode não revelar discrepâncias significativas entre percepções e realidade.


Não Captura a Amplitude : A mídia não captura a amplitude total dos dados, como as variações entre razões de saída individuais.




2. Gráfico de Árvore:

Vantagens:

Estrutura Hierárquica : Um gráfico de árvore pode representar visualmente a relação hierárquica entre as razões de saída e suas subcategorias, se aplicável.


Análise inicial : Pode permitir uma análise detalhada de como diferentes subcategorias apresentadas para as razões de saída.


2023-09-03 (9).png




Limitações:

Complexidade : Gráficos de árvore podem se tornar complexos rapidamente, especialmente se houver muitas razões de saída e subcategorias. Isso pode dificultar a interpretação.


Menos Comparáveis : Pode ser menos direto comparar as percepções da empresa (HRBP) com os dados da pesquisa de resultados usando um gráfico de árvore.
